```python
def get_companies(driver):
    driver.get(company_url)
    company_div_class = 'company-info'
    companies = driver.find_elements_by_class_name(company_div_class)
    return companies
print('Get the Number of companies')
companies = get_companies(driver)
print(f'Found {len(companies)} companies')
```

```python
contact = soup.find_all("a",{"class":"contact-option"})
contact
```

```python
contact_all = soup.find_all("ul",{"class":"contact-options-container-package-redesign"})
contact_all
```

```python


for item in contact_all:
    soup = BeautifulSoup(str(item), 'html.parser')
    links = soup.find_all('a')

    url = ""
    email = ""
    tele_no = ""

    for link in links:
        href = link['href']

        if href.startswith('http'):
            url = href
        elif href.startswith('mailto:'):
            email = href.replace('mailto:', '')
        elif href.startswith('tel:'):
            tele_no = href.replace('tel:', '')

    # Print the extracted data for the current ul element
    print("URL:", url)
    print("Email:", email)
    print("Telephone Number:", tele_no)
    print()
```

```python
extracted_data = []
for item in contact:
    link_element = item.find('a')  # Find the <a> element within the <ul> element
    if link_element is not None:
        href = link_element['href']  # Extract the value of the 'href' attribute
        extracted_data.append(href)

print(extracted_data)
```
